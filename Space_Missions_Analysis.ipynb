# Exploratory Data Analysis (EDA)
#### What is the distribution of numerical features such as mission cost and launch year?

#### Rockets like Cosmos-3MRB (6 times), Lambda-IV S, Titan IV, and Titan IIID (each 5 times) appear more often, while most other rockets appear only once, showing that the rocket category has a lot of variety.

```python
df.StatusMission.value_counts()
```

 #### What is the distribution of mission costs, and are there any notable outliers?

```python
plt.figure(figsize=(10,5))
sb.histplot(df.Price,color='Red')
plt.title('Distribution of Mission Cost',fontsize=14, fontweight='bold')
plt.xlabel('Price',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

#### Most missions have relatively low costs, with only a few costing much higher

```python
plt.figure(figsize=(10,5))
sb.boxplot(x=df.Price,color='Red')
plt.title('Outliers in Distribution Of Mission Cost')
plt.xlabel('Price', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

#### A few missions have much higher costs than most others, making them easy to see in the distribution

# Temporal Trends
#### How has the number of space launches changed over the Years?

```python
a1=df.Year.value_counts().sort_index()
a1
```

```python
plt.figure(figsize=(12,6))
sb.lineplot(x=a1.index,y=a1.values,color='Red')
plt.title('Number of Space Launches Over Year',fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

#### This line plot shows count of year increased over time,the highest number of launches occurs in year 2018 to 2019.

#### Are there any noticeable trends in mission outcomes over the years?

```python
a2=df.groupby(['Year','StatusMission']).size().reset_index()
a2
```

```python
plt.figure(figsize=(15,5))
sb.barplot(x=a2.Year,y=a2[0],hue=a2.StatusMission,palette='dark:Red')
plt.xticks(rotation=90)
plt.legend(title='Mission Status',bbox_to_anchor=(1.0,1.0))
plt.title('Mission Outcomes Over Year',fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

#### In early years,like 1957 space mission were very likely to fail.Now in 2019 to 2020,it is much more likely to succeed and happen more often.
