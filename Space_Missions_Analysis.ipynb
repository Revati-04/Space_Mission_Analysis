```python
import pandas as pd 
import numpy as np 
import seaborn as sb 
import matplotlib.pyplot as plt 
```

```python
df= pd.read_csv('../DataSets/Space_Corrected.csv')
df
```

```python
df.info()
```

# Data Preperation

```python
df= df.drop('Unnamed: 0.1' , axis=1)
df= df.drop('Unnamed: 0', axis=1)
df
```

#### Identify missing values and handle it

```python
df.isnull().sum()
```

```python
df.columns.unique()
```

```python
df.Price=df.Price.fillna(0)
```

#### Rename Columns if needed

```python
df= df.rename(columns= {'Detail':'Rocket'})
```

```python
df= df.rename(columns= {' Rocket':'Price'})
```

```python
df=df.rename(columns={'Company Name':'CompanyName'})
```

```python
df=df.rename(columns={'Status Mission':'StatusMission'})
```

```python
df=df.rename(columns={'Status Rocket':'StatusRocket'})
```

```python
df
```

#### Covert data types if needed

```python
df.dtypes
```

```python
df.Price.unique()
```

```python
df.Price= df.Price.replace('[,]','',regex=True).astype('float64')
df
```

#### Extract month and year column

```python
df['Year'] = df['Datum'].str[12:16]
```

```python
df['Month'] = df['Datum'].str[4:8]
```

```python
df
```

```python
df.Year= df.Year.astype('int64')
```

 #### What is the distribution of mission costs, and are there any notable outliers?

 # Launch Site Analysis

# Cost Analysis
#### Which companies spend the most?

```python
t1=df.groupby('CompanyName')['Price'].mean().nlargest(10)
t1
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=t1.index,y=t1.values,hue=t1,palette='Reds')
plt.title('Most Spend Companies',fontsize=14, fontweight='bold')
plt.xlabel('Companies')
plt.ylabel('Price')
plt.legend(title='Price',bbox_to_anchor=(1.11,1.0))
plt.xticks(rotation=90)
plt.show()
```

#### Which missions or rockets are the most expensive?

```python
t2=df.groupby('Rocket')['Price'].mean().nlargest(10)
t2
```

```python
plt.figure(figsize=(12,6))
sb.barplot(x=t2.index,y=t2.values,hue=t2,palette='Reds')
plt.title('Most Expensive Rockets',fontsize=14, fontweight='bold')
plt.xlabel('Rockets')
plt.ylabel('Price')
plt.legend(title='Price',bbox_to_anchor=(1.12,1.0))
plt.xticks(rotation=90)
plt.show()
```

# Temporal Trends
#### How has the number of space launches changed over the Years?

```python
a1=df.Year.value_counts().sort_index()
a1
```

```python
plt.figure(figsize=(15,6))
sb.lineplot(x=a1.index,y=a1.values,color='Red')
plt.title('Number of Space Launches Over Year',fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

#### Are there any noticeable trends in mission outcomes over the years?

```python
a2=df.groupby(['Year','StatusMission']).size().reset_index()
a2
```

```python
plt.figure(figsize=(15,5))
sb.barplot(x=a2.Year,y=a2[0],hue=a2.StatusMission,palette='dark:Red')
plt.xticks(rotation=90)
plt.legend(title='Mission Status',bbox_to_anchor=(1.0,1.0))
plt.title('Mission Outcomes Over Year',fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```
